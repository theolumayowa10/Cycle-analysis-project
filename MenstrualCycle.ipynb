##### Boxplots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

main_color = "#4169E1"  
secondary_color = "#FF6B6B"  
background_color = "#F8F9FA"  
grid_color = "#DDDDDD"  

def create_combined_visualization(data, title, xlabel, color=main_color, figsize=(12, 8)):
    """
    Creates a combined visualization with boxplot and histogram
    """
    #
    fig, (ax_box, ax_hist) = plt.subplots(
        2, 1, 
        figsize=figsize, 
        gridspec_kw={'height_ratios': [1, 3]},
        sharex=True
    )
    
  
    fig.patch.set_facecolor(background_color)
    ax_box.set_facecolor(background_color)
    ax_hist.set_facecolor(background_color)
    
    
    min_val = max(1, int(data.min()) - 1)
    max_val = int(data.max()) + 1
    bins = np.arange(min_val, max_val + 1, 1) 
    

    n, bins, patches = ax_hist.hist(data.dropna(), bins=bins, color=color, 
                                   alpha=0.7, edgecolor='white', linewidth=0.5)
    

    boxplot = ax_box.boxplot(data.dropna(), vert=False, patch_artist=True,
                            boxprops=dict(color=color, facecolor='white', linewidth=2),
                            whiskerprops=dict(color=color, linewidth=2),
                            capprops=dict(color=color, linewidth=2),
                            medianprops=dict(color=secondary_color, linewidth=3),
                            flierprops=dict(marker='o', markerfacecolor=color, 
                                           markersize=4, alpha=0.7, markeredgecolor='none'))
    

    stats = {
        'median': np.median(data.dropna()),
        'q1': np.percentile(data.dropna(), 25),
        'q3': np.percentile(data.dropna(), 75),
        'mean': np.mean(data.dropna())
    }
    
    textstr = f'n = {len(data.dropna()):,}\nMedian = {stats["median"]:.1f} days\nMean = {stats["mean"]:.1f} days'
    ax_hist.text(0.95, 0.95, textstr, transform=ax_hist.transAxes, fontsize=10,
                verticalalignment='top', horizontalalignment='right',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.9))
    

    fig.suptitle(title, fontsize=16, fontweight='bold', y=0.95)
    ax_hist.set_xlabel(xlabel, fontsize=12, labelpad=10)
    ax_hist.set_ylabel('Frequency', fontsize=12, labelpad=10)
    
    ax_hist.grid(True, axis='y', linestyle='--', alpha=0.7, color=grid_color)
    ax_hist.set_axisbelow(True)
    
    for ax in [ax_box, ax_hist]:
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
    
    ax_box.set_yticks([])
    ax_box.spines['left'].set_visible(False)
    ax_box.spines['bottom'].set_visible(False)
    
    x_ticks = np.arange(0, max_val + 5, 5)
    ax_hist.set_xticks(x_ticks)
    ax_hist.set_xlim(min_val - 1, max_val)
    
    plt.tight_layout()
    return fig, (ax_box, ax_hist)

np.random.seed(42)
cycle_data = pd.Series(np.random.normal(29, 5, 1665))
cycle_data = cycle_data.clip(lower=15, upper=50) 
cycle_data = cycle_data.round().astype(int)  
fig, (ax_box, ax_hist) = create_combined_visualization(
    cycle_data, 
    "Distribution of Menstrual Cycle Lengths", 
    "Cycle Length (days)"
)

plt.show()
```

##### Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scipy.stats as stats
from matplotlib.colors import LinearSegmentedColormap

# Set up the style parameters
main_color = "#4169E1"  
secondary_color = "#FF6B6B"  
background_color = "#F8F9FA"  
grid_color = "#DDDDDD"  

def create_enhanced_correlation_heatmap(df, columns, title, figsize=(12, 10)):
    """
    Creates an enhanced correlation heatmap with significance indicators and better styling
    """
    # Create figure and axis
    fig, ax = plt.subplots(figsize=figsize)
    
    # Set background colors
    fig.patch.set_facecolor(background_color)
    ax.set_facecolor(background_color)
    
    # Calculate correlation matrix
    corr_matrix = df[columns].corr()
    
    # Calculate p-values for correlations
    p_values = np.zeros_like(corr_matrix)
    for i in range(len(columns)):
        for j in range(len(columns)):
            if i != j:
                # Remove NaN values for correlation test
                clean_data = df[[columns[i], columns[j]]].dropna()
                if len(clean_data) > 2:
                    corr, p_val = stats.pearsonr(clean_data[columns[i]], clean_data[columns[j]])
                    p_values[i, j] = p_val
                else:
                    p_values[i, j] = np.nan
            else:
                p_values[i, j] = 0  # Diagonal
    
    # Create custom colormap
    cmap = LinearSegmentedColormap.from_list('custom_blue', ['#FFFFFF', main_color], N=100)
    
    # Create heatmap
    heatmap = sns.heatmap(corr_matrix, annot=True, cmap=cmap, fmt=".2f",
                         center=0, vmin=-1, vmax=1, square=True,
                         cbar_kws={"shrink": .8, "label": "Correlation Coefficient"},
                         ax=ax)
    
    # Add significance stars
    for i in range(len(columns)):
        for j in range(len(columns)):
            if i != j and not np.isnan(p_values[i, j]):
                if p_values[i, j] < 0.001:
                    star = "***"
                elif p_values[i, j] < 0.01:
                    star = "**"
                elif p_values[i, j] < 0.05:
                    star = "*"
                else:
                    star = ""
                
                if star:
                    ax.text(j + 0.5, i + 0.5, star, 
                           ha="center", va="center", 
                           fontsize=10, fontweight="bold",
                           color="white" if abs(corr_matrix.iloc[i, j]) > 0.5 else "black")
    
    # Add title
    ax.set_title(title, fontsize=16, fontweight="bold", pad=20)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)
    
    # Add sample size information
    n_text = f"Sample size: {len(df.dropna(subset=columns)):,}"
    ax.text(0.02, -0.1, n_text, transform=ax.transAxes, fontsize=10,
            bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))
    
    # Add significance legend
    legend_text = "Significance: *p<0.05, **p<0.01, ***p<0.001"
    ax.text(0.02, -0.15, legend_text, transform=ax.transAxes, fontsize=9,
            bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))
    
    plt.tight_layout()
    return fig, ax, corr_matrix, p_values

# Sample data generation for demonstration
np.random.seed(42)
n_samples = 200

# Generate sample data
data = {
    'LengthofCycle': np.random.normal(29, 5, n_samples),
    'MeanCycleLength': np.random.normal(28, 4, n_samples),
    'LengthofLutealPhase': np.random.normal(14, 2, n_samples),
    'LengthofMenses': np.random.normal(5, 1.5, n_samples),
    'TotalFertilityFormula': np.random.normal(75, 15, n_samples),
    'Age': np.random.normal(32, 8, n_samples),
    'BMI': np.random.normal(24, 4, n_samples)
}

# Add some correlations
data['LengthofCycle'] = data['MeanCycleLength'] + np.random.normal(0, 1, n_samples)
data['LengthofLutealPhase'] = 0.3 * data['LengthofCycle'] + np.random.normal(10, 1.5, n_samples)
data['TotalFertilityFormula'] = 80 - 0.5 * data['Age'] + np.random.normal(0, 10, n_samples)

df = pd.DataFrame(data)

# Define columns for correlation analysis
num_cols = ['LengthofCycle', 'MeanCycleLength', 'LengthofLutealPhase', 
            'LengthofMenses', 'TotalFertilityFormula', 'Age', 'BMI']

# Create the enhanced heatmap
fig, ax, corr_matrix, p_values = create_enhanced_correlation_heatmap(
    df, num_cols, 'Correlation Heatmap of Key Menstrual Health Variables'
)

plt.show()
```

##### Relationship Between Cycle length and luteal phase

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scipy.stats as stats


main_color = "#4169E1"  
secondary_color = "#FF6B6B"  
background_color = "#F8F9FA"  
grid_color = "#DDDDDD"  

def create_enhanced_cycle_luteal_scatter(df, x_col, y_col, title, xlabel, ylabel, 
                                       color=main_color, figsize=(10, 7)):
    """
    Creates an enhanced scatterplot with regression line, confidence interval,
    and statistical annotations for cycle length vs luteal phase
    """
   
    fig, ax = plt.subplots(figsize=figsize)
    
   
    fig.patch.set_facecolor(background_color)
    ax.set_facecolor(background_color)
    
   
    clean_df = df.dropna(subset=[x_col, y_col])
    x_values = clean_df[x_col].values
    y_values = clean_df[y_col].values
    
   
    scatter = ax.scatter(x_values, y_values, color=color, alpha=0.6, 
                        s=40, edgecolors='white', linewidth=0.5)
    
    
    if len(x_values) > 1:
        sns.regplot(x=x_col, y=y_col, data=clean_df, scatter=False, 
                   line_kws={'color': secondary_color, 'linewidth': 2.5}, 
                   ci=95, ax=ax)
    
    
    if len(x_values) > 1:
       
        r_pearson, p_pearson = stats.pearsonr(x_values, y_values)
        
        
        r_spearman, p_spearman = stats.spearmanr(x_values, y_values)
        
   
        slope, intercept, r_value, p_value, std_err = stats.linregress(x_values, y_values)
        
        
        ax.axhline(y=14, color='gray', linestyle='--', alpha=0.7, linewidth=1.5,
                  label='Typical Luteal Phase (12-16 days)')
        
       
        stats_text = (f'Pearson r = {r_pearson:.3f} (p = {p_pearson:.4f})\n'
                     f'Spearman ρ = {r_spearman:.3f} (p = {p_spearman:.4f})\n'
                     f'R² = {r_value**2:.3f}\n'
                     f'y = {slope:.3f}x + {intercept:.3f}')
        
        ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, fontsize=10,
                verticalalignment='top', horizontalalignment='left',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.9))
    
   
    ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
    ax.set_xlabel(xlabel, fontsize=12, labelpad=10)
    ax.set_ylabel(ylabel, fontsize=12, labelpad=10)
    
  
    ax.grid(True, linestyle='--', alpha=0.7, color=grid_color)
    ax.set_axisbelow(True)
    
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
   
    n_text = f'n = {len(clean_df):,}'
    ax.text(0.98, 0.02, n_text, transform=ax.transAxes, fontsize=10,
            verticalalignment='bottom', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
   
    ax.legend(loc='lower right', frameon=True, facecolor='white')
    
    
    ax.set_xlim(max(15, x_values.min() - 2), min(50, x_values.max() + 2))
    ax.set_ylim(max(5, y_values.min() - 1), min(20, y_values.max() + 1))
    
    plt.tight_layout()
    return fig, ax


np.random.seed(42)
n_samples = 200


cycle_lengths = np.random.normal(29, 5, n_samples)
cycle_lengths = np.clip(cycle_lengths, 21, 45)  # Realistic range


luteal_lengths = np.random.normal(14, 2, n_samples) + 0.1 * (cycle_lengths - 29)
luteal_lengths = np.clip(luteal_lengths, 10, 18)  # Clinical range

df = pd.DataFrame({
    'LengthofCycle': cycle_lengths,
    'LengthofLutealPhase': luteal_lengths
})


fig, ax = create_enhanced_cycle_luteal_scatter(
    df, 
    x_col='LengthofCycle', 
    y_col='LengthofLutealPhase', 
    title='Relationship Between Cycle Length and Luteal Phase Duration',
    xlabel='Cycle Length (days)',
    ylabel='Luteal Phase Duration (days)'
)

plt.show()
```

##### Phase Comparison

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scipy.stats as stats


main_color = "#4169E1"  
secondary_color = "#FF6B6B"  
background_color = "#F8F9FA"  
grid_color = "#DDDDDD"  

def create_enhanced_phase_boxplot(df, x_col, y_col, title, xlabel, ylabel, 
                                 color=main_color, figsize=(10, 6)):
    """
    Creates an enhanced boxplot with statistical annotations and better styling
    """
    
    fig, ax = plt.subplots(figsize=figsize)
    
   
    fig.patch.set_facecolor(background_color)
    ax.set_facecolor(background_color)
    
    
    phases = sorted(df[x_col].unique())
    
    
    boxplot = sns.boxplot(x=x_col, y=y_col, data=df, ax=ax, color=color,
                         fliersize=3, linewidth=1.5,
                         boxprops=dict(alpha=0.8, edgecolor='black'),
                         whiskerprops=dict(color='black', linewidth=1.5),
                         capprops=dict(color='black', linewidth=1.5),
                         medianprops=dict(color=secondary_color, linewidth=2))
    
    
    sns.stripplot(x=x_col, y=y_col, data=df, ax=ax, color=color, 
                 alpha=0.4, size=3, jitter=True)
    
    
    for i, phase in enumerate(phases):
        n = len(df[df[x_col] == phase].dropna(subset=[y_col]))
        ax.text(i, ax.get_ylim()[1] * 0.95, f'n={n}', 
                horizontalalignment='center', fontsize=10,
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    
    ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
    ax.set_xlabel(xlabel, fontsize=12, labelpad=10)
    ax.set_ylabel(ylabel, fontsize=12, labelpad=10)
    
   
    ax.grid(True, axis='y', linestyle='--', alpha=0.7, color=grid_color)
    ax.set_axisbelow(True)
    
   
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
   
    if len(phases) > 1:
       
        groups = [df[df[x_col] == phase][y_col].dropna().values for phase in phases]
        
       
        if all(len(group) >= 2 for group in groups):
            try:
                
                stat, p_value = stats.kruskal(*groups)
                
                
                p_text = f'Kruskal-Wallis p = {p_value:.4f}' if p_value >= 0.0001 else 'Kruskal-Wallis p < 0.0001'
                ax.text(0.02, 0.98, p_text, transform=ax.transAxes, fontsize=10,
                        verticalalignment='top', horizontalalignment='left',
                        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
                
               
                if len(phases) <= 5:
                   
                    pairs = []
                    p_values = []
                    
                    for i in range(len(phases)):
                        for j in range(i+1, len(phases)):
                            if len(groups[i]) >= 2 and len(groups[j]) >= 2:
                               
                                try:
                                    _, p_val = stats.mannwhitneyu(groups[i], groups[j], alternative='two-sided')
                                    pairs.append((i, j))
                                    p_values.append(p_val)
                                except:
                                    continue
                    
                    
                    if pairs:
                        y_max = ax.get_ylim()[1]
                        y_range = ax.get_ylim()[1] - ax.get_ylim()[0]
                        
                        for idx, (pair, p_val) in enumerate(zip(pairs, p_values)):
                            if p_val < 0.05:  
                                y_pos = y_max + 0.05 * y_range * (idx + 1)
                                
                                
                                ax.plot([pair[0], pair[0], pair[1], pair[1]], 
                                       [y_pos, y_pos+0.02*y_range, y_pos+0.02*y_range, y_pos], 
                                       lw=1.5, color='black')
                                
                                
                                if p_val < 0.001:
                                    stars = '***'
                                elif p_val < 0.01:
                                    stars = '**'
                                elif p_val < 0.05:
                                    stars = '*'
                                else:
                                    stars = ''
                                
                                if stars:  
                                    ax.text((pair[0] + pair[1]) / 2, y_pos + 0.03*y_range, stars, 
                                           ha='center', va='bottom', color='black', fontsize=12, fontweight='bold')
                        
                        
                        if any(p_val < 0.05 for p_val in p_values):
                            significant_pairs = sum(1 for p_val in p_values if p_val < 0.05)
                            ax.set_ylim(ax.get_ylim()[0], y_max + 0.1 * y_range * (significant_pairs + 1))
            except Exception as e:
                print(f"Statistical analysis skipped due to error: {e}")
    
    plt.tight_layout()
    return fig, ax


np.random.seed(42)
n_samples = 200


phases = ['Light', 'Medium', 'Heavy']
phase_probs = [0.4, 0.4, 0.2]  


bleeding_phases = np.random.choice(phases, n_samples, p=phase_probs)


menses_lengths = []
for phase in bleeding_phases:
    if phase == 'Light':
        length = np.random.normal(3, 1)
    elif phase == 'Medium':
        length = np.random.normal(5, 1.5)
    else:  # Heavy
        length = np.random.normal(7, 2)
    menses_lengths.append(max(1, length)) 

df = pd.DataFrame({
    'PhasesBleeding': bleeding_phases,
    'LengthofMenses': menses_lengths
})


fig, ax = create_enhanced_phase_boxplot(
    df, 
    x_col='PhasesBleeding', 
    y_col='LengthofMenses', 
    title='Length of Menses Across Bleeding Phases',
    xlabel='Bleeding Phase',
    ylabel='Length of Menses (days)'
)

plt.show()
```

##### Relationship Between Age and Mean Cycle Length

```python
main_color = "#4169E1"  
secondary_color = "#FF6B6B"  
background_color = "#F8F9FA"  
grid_color = "#DDDDDD"  

def create_enhanced_scatterplot(df, x_col, y_col, title, xlabel, ylabel, 
                               color=main_color, figsize=(10, 7)):
    """
    Creates an enhanced scatterplot with regression line, confidence interval,
    and statistical annotations
    """
  
    fig, ax = plt.subplots(figsize=figsize)
    

    fig.patch.set_facecolor(background_color)
    ax.set_facecolor(background_color)
    
  
    x_values = df[x_col].values
    y_values = df[y_col].values
    valid_indices = ~(np.isnan(x_values) | np.isnan(y_values))
    x_clean = x_values[valid_indices]
    y_clean = y_values[valid_indices]
    
    if len(x_clean) > 1:
        slope, intercept, r_value, p_value, std_err = stats.linregress(x_clean, y_clean)
        regression_line = slope * x_clean + intercept
        equation_text = f'y = {slope:.2f}x + {intercept:.2f}'
        r_squared_text = f'R² = {r_value**2:.3f}'
        p_value_text = f'p = {p_value:.4f}' if p_value >= 0.0001 else 'p < 0.0001'
    else:
        slope, intercept, r_value, p_value = 0, 0, 0, 1
        equation_text = 'Insufficient data for regression'
        r_squared_text = ''
        p_value_text = ''
    

    scatter = ax.scatter(df[x_col], df[y_col], color=color, alpha=0.6, 
                        s=40, edgecolors='white', linewidth=0.5)
    
    if len(x_clean) > 1:
        sns.regplot(x=x_col, y=y_col, data=df, scatter=False, 
                   line_kws={'color': secondary_color, 'linewidth': 2.5}, ax=ax)
    

    ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
    ax.set_xlabel(xlabel, fontsize=12, labelpad=10)
    ax.set_ylabel(ylabel, fontsize=12, labelpad=10)
    
    
    ax.grid(True, linestyle='--', alpha=0.7, color=grid_color)
    ax.set_axisbelow(True)
    

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    

    if len(x_clean) > 1:
        annotation_text = f'{equation_text}\n{r_squared_text}\n{p_value_text}'
        ax.text(0.02, 0.98, annotation_text, transform=ax.transAxes, fontsize=10,
                verticalalignment='top', horizontalalignment='left',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
   
    n_text = f'n = {len(df.dropna(subset=[x_col, y_col]))}'
    ax.text(0.98, 0.02, n_text, transform=ax.transAxes, fontsize=10,
            verticalalignment='bottom', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    
    min_age = max(0, int(df[x_col].min()) - 1)
    max_age = int(df[x_col].max()) + 1
    ax.set_xlim(min_age, max_age)
    
    
    min_cycle = max(0, int(df[y_col].min()) - 2)
    max_cycle = int(df[y_col].max()) + 2
    ax.set_ylim(min_cycle, max_cycle)
    
    plt.tight_layout()
    return fig, ax


np.random.seed(42)
n_samples = 200
ages = np.random.randint(18, 45, n_samples)

mean_cycle_length = 29 - 0.08 * (ages - 25) + np.random.normal(0, 2, n_samples)
mean_cycle_length = np.clip(mean_cycle_length, 21, 35)  
df = pd.DataFrame({'Age': ages, 'MeanCycleLength': mean_cycle_length})


fig, ax = create_enhanced_scatterplot(
    df, 
    x_col='Age', 
    y_col='MeanCycleLength', 
    title='Relationship Between Age and Mean Cycle Length',
    xlabel='Age (years)',
    ylabel='Mean Cycle Length (days)'
)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from tabulate import tabulate
from matplotlib.gridspec import GridSpec
import scipy.stats as stats

# Set up the style parameters
main_color = "#4169E1"  
secondary_color = "#FF6B6B"  
background_color = "#F8F9FA"  
grid_color = "#DDDDDD"  

def analyze_extreme_cycles(df, output_type='print'):
    """
    Analyzes and visualizes extreme cycle lengths with enhanced reporting
    """
    # Define extreme thresholds (clinical guidelines)
    short_threshold = 21
    long_threshold = 35
    
    # Identify extreme cycles
    extreme_cycles = df[(df['LengthofCycle'] < short_threshold) | (df['LengthofCycle'] > long_threshold)].copy()
    extreme_cycles['Cycle_Type'] = np.where(extreme_cycles['LengthofCycle'] < short_threshold, 'Short', 'Long')
    
    if output_type == 'print':
        print("="*80)
        print("EXTREME CYCLE LENGTH ANALYSIS")
        print("="*80)
        
        # Summary statistics
        total_cycles = len(df.dropna(subset=['LengthofCycle']))
        extreme_count = len(extreme_cycles)
        extreme_percent = (extreme_count / total_cycles * 100) if total_cycles > 0 else 0
        
        print(f"\nTotal cycles analyzed: {total_cycles:,}")
        print(f"Extreme cycles found: {extreme_count:,} ({extreme_percent:.1f}%)")
        print(f"Short cycles (<{short_threshold} days): {len(extreme_cycles[extreme_cycles['Cycle_Type'] == 'Short']):,}")
        print(f"Long cycles (>{long_threshold} days): {len(extreme_cycles[extreme_cycles['Cycle_Type'] == 'Long']):,}")
        
        # Detailed table
        if not extreme_cycles.empty:
            print("\nDetailed Extreme Cycle Cases:")
            table_data = extreme_cycles[['ClientID', 'LengthofCycle', 'Cycle_Type', 'MeanCycleLength', 'Age']]
            table_data = table_data.sort_values('LengthofCycle')
            print(tabulate(table_data, headers='keys', tablefmt='grid', showindex=False))
        else:
            print("\nNo extreme cycles found.")
            
    return extreme_cycles

def analyze_extreme_luteal_phases(df, output_type='print'):
    """
    Analyzes and visualizes extreme luteal phase lengths with enhanced reporting
    """
    # Clinical threshold for long luteal phase
    luteal_threshold = 18
    
    # Identify extreme luteal phases
    extreme_luteal = df[df['LengthofLutealPhase'] > luteal_threshold].copy()
    
    if output_type == 'print':
        print("\n" + "="*80)
        print("EXTREME LUTEAL PHASE ANALYSIS")
        print("="*80)
        
        # Summary statistics
        total_luteal = len(df.dropna(subset=['LengthofLutealPhase']))
        extreme_count = len(extreme_luteal)
        extreme_percent = (extreme_count / total_luteal * 100) if total_luteal > 0 else 0
        
        print(f"\nTotal luteal phases analyzed: {total_luteal:,}")
        print(f"Extreme luteal phases (> {luteal_threshold} days): {extreme_count:,} ({extreme_percent:.1f}%)")
        
        # Detailed table
        if not extreme_luteal.empty:
            print("\nDetailed Extreme Luteal Phase Cases:")
            table_data = extreme_luteal[['ClientID', 'LengthofLutealPhase', 'LengthofCycle', 'Age']]
            table_data = table_data.sort_values('LengthofLutealPhase', ascending=False)
            print(tabulate(table_data, headers='keys', tablefmt='grid', showindex=False))
        else:
            print("\nNo extreme luteal phases found.")
            
    return extreme_luteal

def create_extreme_values_visualization(df, figsize=(15, 10)):
    """
    Creates comprehensive visualization of extreme values
    """
    fig = plt.figure(figsize=figsize)
    fig.patch.set_facecolor(background_color)
    
    # Create grid layout
    gs = GridSpec(3, 2, figure=fig)
    
    # 1. Extreme cycle lengths distribution
    ax1 = fig.add_subplot(gs[0, 0])
    short_cycles = df[df['LengthofCycle'] < 21]
    long_cycles = df[df['LengthofCycle'] > 35]
    
    bins = np.arange(15, 50, 2)
    ax1.hist(df['LengthofCycle'].dropna(), bins=bins, color=main_color, alpha=0.7, 
             label='Normal Cycles', edgecolor='white')
    ax1.hist(short_cycles['LengthofCycle'], bins=bins, color=secondary_color, alpha=0.8,
             label='Short Cycles (<21 days)')
    ax1.hist(long_cycles['LengthofCycle'], bins=bins, color='orange', alpha=0.8,
             label='Long Cycles (>35 days)')
    
    ax1.axvline(21, color=secondary_color, linestyle='--', alpha=0.8, linewidth=2)
    ax1.axvline(35, color='orange', linestyle='--', alpha=0.8, linewidth=2)
    ax1.set_xlabel('Cycle Length (days)')
    ax1.set_ylabel('Frequency')
    ax1.set_title('Distribution of Cycle Lengths with Extreme Values')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Extreme luteal phases distribution
    ax2 = fig.add_subplot(gs[0, 1])
    extreme_luteal = df[df['LengthofLutealPhase'] > 18]
    
    bins = np.arange(8, 25, 1)
    ax2.hist(df['LengthofLutealPhase'].dropna(), bins=bins, color=main_color, alpha=0.7,
             label='Normal Luteal Phase', edgecolor='white')
    ax2.hist(extreme_luteal['LengthofLutealPhase'], bins=bins, color=secondary_color, alpha=0.8,
             label='Long Luteal Phase (>18 days)')
    
    ax2.axvline(18, color=secondary_color, linestyle='--', alpha=0.8, linewidth=2)
    ax2.set_xlabel('Luteal Phase Length (days)')
    ax2.set_ylabel('Frequency')
    ax2.set_title('Distribution of Luteal Phase Lengths')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Age distribution of extreme cases - FIXED VERSION
    ax3 = fig.add_subplot(gs[1, 0])
    extreme_cases = pd.concat([
        df[df['LengthofCycle'] < 21].assign(Type='Short Cycle'),
        df[df['LengthofCycle'] > 35].assign(Type='Long Cycle'),
        df[df['LengthofLutealPhase'] > 18].assign(Type='Long Luteal')
    ])
    
    if not extreme_cases.empty:
        # Create custom colors for each type
        colors = [secondary_color, 'orange', 'purple']
        types = extreme_cases['Type'].unique()
        
        # Create boxplot without palette warning
        box_data = []
        for i, type_name in enumerate(types):
            type_data = extreme_cases[extreme_cases['Type'] == type_name]['Age'].dropna()
            if len(type_data) > 0:
                box_data.append(type_data)
        
        if box_data:
            box_plot = ax3.boxplot(box_data, patch_artist=True)
            
            # Set colors for boxes
            for i, box in enumerate(box_plot['boxes']):
                box.set(facecolor=colors[i], alpha=0.7)
            
            # Set median line color
            for median in box_plot['medians']:
                median.set(color='black', linewidth=2)
            
            ax3.set_xticks(range(1, len(types) + 1))
            ax3.set_xticklabels(types, rotation=45, ha='right')
            ax3.set_title('Age Distribution of Extreme Cases')
            ax3.set_ylabel('Age')
            ax3.grid(True, alpha=0.3)
    
    # 4. Correlation between extreme cycle and luteal phase
    ax4 = fig.add_subplot(gs[1, 1])
    extreme_both = df[((df['LengthofCycle'] < 21) | (df['LengthofCycle'] > 35)) & 
                     (df['LengthofLutealPhase'] > 18)]
    
    if not extreme_both.empty:
        ax4.scatter(extreme_both['LengthofCycle'], extreme_both['LengthofLutealPhase'], 
                   color=secondary_color, alpha=0.7, s=50)
        ax4.set_xlabel('Cycle Length (days)')
        ax4.set_ylabel('Luteal Phase Length (days)')
        ax4.set_title('Extreme Cases: Cycle vs Luteal Phase Length')
        ax4.axvline(21, color=secondary_color, linestyle='--', alpha=0.5)
        ax4.axvline(35, color='orange', linestyle='--', alpha=0.5)
        ax4.axhline(18, color='purple', linestyle='--', alpha=0.5)
        ax4.grid(True, alpha=0.3)
    else:
        ax4.text(0.5, 0.5, 'No cases with both extreme cycle\nand extreme luteal phase', 
                ha='center', va='center', transform=ax4.transAxes)
        ax4.set_xticks([])
        ax4.set_yticks([])
    
    # 5. Summary statistics
    ax5 = fig.add_subplot(gs[2, :])
    ax5.axis('off')
    
    # Calculate summary statistics
    total_cycles = len(df.dropna(subset=['LengthofCycle']))
    short_cycles_count = len(df[df['LengthofCycle'] < 21])
    long_cycles_count = len(df[df['LengthofCycle'] > 35])
    extreme_luteal_count = len(df[df['LengthofLutealPhase'] > 18])
    
    summary_text = (
        f"EXTREME VALUES SUMMARY\n\n"
        f"Total cycles analyzed: {total_cycles:,}\n"
        f"Short cycles (<21 days): {short_cycles_count:,} ({short_cycles_count/total_cycles*100:.1f}%)\n"
        f"Long cycles (>35 days): {long_cycles_count:,} ({long_cycles_count/total_cycles*100:.1f}%)\n"
        f"Long luteal phases (>18 days): {extreme_luteal_count:,}\n\n"
        f"Clinical Notes:\n"
        f"- Short cycles may indicate luteal phase deficiency\n"
        f"- Long cycles may suggest anovulation or PCOS\n"
        f"- Long luteal phases may indicate progesterone issues"
    )
    
    ax5.text(0.05, 0.95, summary_text, transform=ax5.transAxes, fontsize=11,
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.9))
    
    plt.tight_layout()
    return fig

def generate_extreme_cases_report(df, filename=None):
    """
    Generates a comprehensive report of extreme cases
    """
    # Analyze extreme cycles
    extreme_cycles = analyze_extreme_cycles(df, output_type='return')
    
    # Analyze extreme luteal phases
    extreme_luteal = analyze_extreme_luteal_phases(df, output_type='return')
    
    # Create visualization
    fig = create_extreme_values_visualization(df)
    
    # Save report if filename provided
    if filename:
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"\nReport saved as {filename}")
    
    return {
        'extreme_cycles': extreme_cycles,
        'extreme_luteal': extreme_luteal,
        'figure': fig
    }

# Sample data generation for demonstration
np.random.seed(42)
n_samples = 500

# Generate sample data
data = {
    'ClientID': [f'CLIENT_{i:03d}' for i in range(n_samples)],
    'LengthofCycle': np.random.normal(29, 6, n_samples),
    'MeanCycleLength': np.random.normal(28, 4, n_samples),
    'LengthofLutealPhase': np.random.normal(14, 3, n_samples),
    'Age': np.random.normal(32, 8, n_samples),
    'BMI': np.random.normal(24, 4, n_samples)
}

# Add some extreme cases
data['LengthofCycle'][:15] = np.random.uniform(15, 20, 15)  # Short cycles
data['LengthofCycle'][15:30] = np.random.uniform(36, 45, 15)  # Long cycles
data['LengthofLutealPhase'][30:40] = np.random.uniform(19, 25, 10)  # Long luteal phases

df = pd.DataFrame(data)

# Generate comprehensive report
report = generate_extreme_cases_report(df)

# Show the visualization
plt.show()

# Print detailed analysis
analyze_extreme_cycles(df, output_type='print')
analyze_extreme_luteal_phases(df, output_type='print')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
from matplotlib.gridspec import GridSpec


main_color = "#4169E1"  
secondary_color = "#FF6B6B"  
background_color = "#F8F9FA"  
grid_color = "#DDDDDD"  

def create_enhanced_pairplot(df, corr_cols, figsize=(14, 12)):
    """
    Creates an enhanced pairplot with correlation coefficients and significance indicators
    """
    
    corr_matrix = df[corr_cols].corr()
    p_values = np.zeros_like(corr_matrix)
    
    for i in range(len(corr_cols)):
        for j in range(len(corr_cols)):
            if i != j:
                clean_data = df[[corr_cols[i], corr_cols[j]]].dropna()
                if len(clean_data) > 2:
                    _, p_val = stats.pearsonr(clean_data[corr_cols[i]], clean_data[corr_cols[j]])
                    p_values[i, j] = p_val
                else:
                    p_values[i, j] = np.nan
    
   
    g = sns.pairplot(df[corr_cols], 
                    diag_kind="kde", 
                    plot_kws={
                        "color": main_color, 
                        "alpha": 0.6,
                        "s": 40,
                        "edgecolor": "white",
                        "linewidth": 0.5
                    },
                    diag_kws={
                        "color": main_color,
                        "fill": True,
                        "alpha": 0.7,
                        "linewidth": 2
                    })
    
    
    g.fig.set_facecolor(background_color)
    for ax in g.axes.flatten():
        ax.set_facecolor(background_color)
    
    
    for i in range(len(corr_cols)):
        for j in range(len(corr_cols)):
            if i < j:  # Upper triangle
                ax = g.axes[i, j]
                corr_val = corr_matrix.iloc[i, j]
                p_val = p_values[i, j]
                
                
                if p_val < 0.001:
                    stars = "***"
                elif p_val < 0.01:
                    stars = "**"
                elif p_val < 0.05:
                    stars = "*"
                else:
                    stars = ""
                
              
                ax.text(0.05, 0.95, f"r = {corr_val:.2f}{stars}", 
                       transform=ax.transAxes, 
                       fontsize=10,
                       fontweight='bold',
                       bbox=dict(boxstyle="round", facecolor="white", alpha=0.9),
                       verticalalignment='top')
    
   
    n_samples = len(df.dropna(subset=corr_cols))
    g.fig.text(0.02, 0.02, f"Sample size: {n_samples:,}", 
              fontsize=10,
              bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))
    
   
    legend_text = "Significance: *p<0.05, **p<0.01, ***p<0.001"
    g.fig.text(0.02, 0.05, legend_text, 
              fontsize=9,
              bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))
    
    g.fig.suptitle("Enhanced Pairplot of Menstrual Cycle Variables\nwith Correlation Coefficients", 
                  fontsize=16, fontweight='bold', y=1.02)
    
    plt.tight_layout()
    return g

def create_focused_pairplot(df, x_vars, y_var, figsize=(12, 10)):
    """
    Creates a focused pairplot for a specific target variable
    """
    
    n_vars = len(x_vars)
    fig, axes = plt.subplots(n_vars, n_vars, figsize=figsize)
    fig.patch.set_facecolor(background_color)
    
    
    axes_flat = axes.flatten()
    
    
    plot_idx = 0
    for i, x_var in enumerate(x_vars):
        for j, y_var_temp in enumerate(x_vars):
            ax = axes[i, j]
            ax.set_facecolor(background_color)
            
            if i == j: 
                data = df[x_var].dropna()
                if len(data) > 0:
                    ax.hist(data, bins=20, color=main_color, alpha=0.7, 
                           edgecolor='white', density=True)
                    ax.set_xlabel(x_var)
                    ax.set_ylabel('Density')
                else:
                    ax.text(0.5, 0.5, 'No data', transform=ax.transAxes, 
                           ha='center', va='center')
            
            elif i > j: 
                clean_data = df[[x_var, y_var_temp]].dropna()
                if len(clean_data) > 0:
                    ax.scatter(clean_data[x_var], clean_data[y_var_temp], 
                              color=main_color, alpha=0.6, s=30)
                    ax.set_xlabel(x_var)
                    ax.set_ylabel(y_var_temp)
                    
                    
                    if len(clean_data) > 2:
                        corr_val = clean_data[x_var].corr(clean_data[y_var_temp])
                        ax.text(0.05, 0.95, f"r = {corr_val:.2f}", 
                               transform=ax.transAxes, fontsize=9,
                               bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))
                else:
                    ax.text(0.5, 0.5, 'No data', transform=ax.transAxes, 
                           ha='center', va='center')
            
            else:  
                ax.axis('off')
    
    plt.suptitle(f"Focused Pairplot for {y_var}", fontsize=16, fontweight='bold', y=0.95)
    plt.tight_layout()
    return fig

def create_pairplot_with_marginals(df, x_vars, y_var, figsize=(15, 12)):
    """
    Creates a pairplot with marginal distributions and regression lines
    """
    fig = plt.figure(figsize=figsize)
    fig.patch.set_facecolor(background_color)
    
    n_vars = len(x_vars)
    gs = GridSpec(n_vars + 1, n_vars + 1, figure=fig)
    
  
    for i, x_var in enumerate(x_vars):
        for j in range(n_vars):
            if i != j:
                ax = fig.add_subplot(gs[i, j])
                ax.set_facecolor(background_color)
                
                clean_data = df[[x_vars[i], x_vars[j]]].dropna()
                if len(clean_data) > 0:
                    ax.scatter(clean_data[x_vars[i]], clean_data[x_vars[j]], 
                              color=main_color, alpha=0.6, s=30)
                    
                    
                    if len(clean_data) > 2:
                        sns.regplot(x=x_vars[i], y=x_vars[j], data=clean_data, 
                                   scatter=False, ax=ax, color=secondary_color,
                                   line_kws={'linewidth': 2})
                
                ax.grid(True, alpha=0.3, linestyle='--')
    
    
    for i in range(n_vars):
        ax = fig.add_subplot(gs[i, i])
        ax.set_facecolor(background_color)
        
        data = df[x_vars[i]].dropna()
        if len(data) > 0:
            ax.hist(data, bins=20, color=main_color, alpha=0.7, 
                   edgecolor='white', density=True)
            ax.grid(True, alpha=0.3, linestyle='--')
    
    plt.suptitle("Advanced Pairplot with Marginal Distributions", 
                fontsize=16, fontweight='bold', y=0.95)
    plt.tight_layout()
    return fig


np.random.seed(42)
n_samples = 300


data = {
    'LengthofCycle': np.random.normal(29, 5, n_samples),
    'MeanCycleLength': np.random.normal(28, 4, n_samples),
    'LengthofLutealPhase': np.random.normal(14, 2, n_samples),
    'LengthofMenses': np.random.normal(5, 1.5, n_samples),
    'TotalFertilityFormula': np.random.normal(75, 15, n_samples)
}


data['LengthofCycle'] = data['MeanCycleLength'] + np.random.normal(0, 1, n_samples)
data['LengthofLutealPhase'] = 0.4 * data['LengthofCycle'] + np.random.normal(8, 1.5, n_samples)
data['TotalFertilityFormula'] = 80 - 0.3 * data['LengthofCycle'] + np.random.normal(0, 10, n_samples)

df = pd.DataFrame(data)


corr_cols = ['LengthofCycle', 'MeanCycleLength', 'LengthofLutealPhase', 'LengthofMenses', 'TotalFertilityFormula']


enhanced_g = create_enhanced_pairplot(df, corr_cols)
plt.show()


predictor_vars = ['LengthofCycle', 'MeanCycleLength', 'LengthofLutealPhase', 'LengthofMenses']
focused_fig = create_focused_pairplot(df, predictor_vars, 'TotalFertilityFormula')
plt.show()


advanced_fig = create_pairplot_with_marginals(df, predictor_vars, 'TotalFertilityFormula')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

# Select only the columns that actually exist
features = df[['LengthofCycle',
               'MeanCycleLength',
               'LengthofLutealPhase',
               'LengthofMenses',
               'TotalFertilityFormula']].dropna()

# Train Isolation Forest (simple anomaly detector)
iso = IsolationForest(contamination=0.05, random_state=42)
features['anomaly'] = iso.fit_predict(features)

# -1 = abnormal cycle, 1 = normal cycle
features['anomaly_label'] = features['anomaly'].map({1: 'Normal', -1: 'Abnormal'})

features.head()
```
